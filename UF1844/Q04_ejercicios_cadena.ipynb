# Ejercicios cadena

```python
# 1. El número total de caracteres de la cadena
cadena = input('Dame una cadena')
print('Numero total de caracteres: ', len(cadena))

# 2. La cadena repetida 10 veces
print(cadena * 10)

# 3. El primer carácter de la cadena (recuerde que los índices de cadena empiezan en 0).
print(cadena[0])

# 4. Los tres primeros caracteres de la cadena
print(cadena[:3])

# 5. Los tres últimos caracteres de la cadena
print(cadena[-3:])

# 6. La cadena al revés
print(cadena[::-1])

# 7. El séptimo carácter de la cadena si la cadena es lo suficientemente larga; en caso contrario, un mensaje
if len(cadena) >= 7:
    print("El séptimo carácter de la cadena es:", cadena[6])
else:
    print("La cadena no es lo suficientemente larga para tener un séptimo carácter")

# 8. La cadena sin el primer y el último carácter
```
