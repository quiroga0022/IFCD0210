```python
def mostrar_config(**kwargs):
    for k,v in kwargs.items():
        print(f'(k): {v}')

mostrar_config(usr='teo', ip='1.2.3.4', puerto='8080', altura=1.77)
```

```python
def mostrar_config(**kwargs):
    USUARIO = kwargs.get('usr','PACO')
    IP = kwargs.get ('ip', '127.0.0.1')
    PUERTO = kwargs.get('puerto','80')

    print(USUARIO, PUERTO, IP)

    mostrar_config(usr='teo', ip='1.2.3.4', altura=1.77)
```

```python
def log(mensaje,*args, **kwargs):
    nivel = kwargs.get('nivel','DEBUG')
    prefijo = kwargs.get('prefijo', '')
    msj_formato = f'[{nivel}]: {prefijo} - {mensaje}'
    return msj_formato

log('Error de configuración', nivel='DEBUG')
```

```python
def http_get(url, **kwargs):
    timeout = kwargs.get('timeout',5)
    headers = kwargs.get('headers', {})
    params = kwargs.get('params',{})

    return {
        'url': url,
        'timeout' : timeout,
        'headers' : headers,
        'params' : params
    }

http_get(
    'https://google.com',
    timeout=10,
    params={'page':2}
)
```
