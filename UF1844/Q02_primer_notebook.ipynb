# Ejercicios de Cadena

```python
# 1. El número total de caracteres de la cadena
cadena = input('Dame una cadena')
print('Numero total de caracteres: ', len(cadena))

# 2. La cadena repetida 10 veces
print(f' {cadena * 10}')

# 3. El primer carácter de la cadena (recuerde que los índices de cadena empiezan en 0).
print(f' {cadena[0]}')

# 4. Los tres primeros caracteres de la cadena
print(f'{cadena[:3]}')

# 5. Los tres últimos caracteres de la cadena
print(f'{cadena[-3:]}')

# 6. La cadena al revés
print(f'{cadena[::-1]}')

# 7. El séptimo carácter de la cadena si la cadena es lo suficientemente larga; en caso contrario, un mensaje
if len(cadena) >= 7:
    print("El séptimo carácter de la cadena es:", cadena[6])
else:
    print("La cadena no es lo suficientemente larga para tener un séptimo carácter")

# 8. La cadena sin el primer y el último carácter.
print(f'(h) {cadena[1:-1]}')

# 9. La cadena en mayúsculas
print(f'(i) {cadena.upper()}')

# 10. Cambiar la letra a por la e.
print(f'(j){cadena.replace('a', 'e')}')

# 11. La cadena con cada letra reemplazada por un espacio
print(f'(k) {'*'.len(cadena)}')

# 2. Una forma sencilla de estimar el número de palabras en una cadena es contar el número de espacios en la cadena.
# Escriba un programa que solicite al usuario una cadena y devuelva una estimación de cuántas palabras tiene. 
cadena = ' Esta es una cadena de prueba '
print(f'{cadena.count(' ') + 1}')

len(cadena.split())

cadena.rsplit

# 9. Solicite al usuario un número e imprima lo siguiente, donde el patrón 
# termina en el número ingresado por el usuario.

numero = int(input('Ingrese un número: '))
for i in range(1, numero + 1):
    print(' ' * (i - 1) + str(i))


```
