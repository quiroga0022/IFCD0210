Un buen programa se asegurará de que los datos que ingresan los usuarios sean válidos.
Escribe un programa que solicite al usuario un peso y lo convierta de kilogramos a libras.
Siempre que el usuario ingrese un peso menor a 0, el programa le indicará que su entrada no es válida y le solicitará nuevamente que ingrese un peso.
[Sugerencia: Usa un bucle while, no una sentencia if].

```python
kg = float(input('Introduce el peso en kilogramos'))
FACTOR = 2.204623
while kg < 0:
    kg = float(input('Dime un peso en kg'))

    print(f'{kg} kilos son {kg * FACTOR} libras.')
    
```

 4 Escribe un programa que solicite al usuario una contraseña.
 Si el usuario ingresa la contraseña correcta, el programa le indicará que ha iniciado sesión en  el sistema.
 De lo contrario, le pedirá que vuelva a ingresar la contraseña.
 El usuario solo tendrá cinco intentos para ingresar la contraseña, tras lo cual el programa le indicará que ha sido expulsado del sistema.

```python
clave = 'manolito'
MAX_INTENTOS = 5
intentos = 0
while intentos < MAX_INTENTOS:
    user_pass = input('Introduzca la clave: ')
    if clave != user_pass:
        print('Clave INCORRECTA')
        intentos += 1
    else: 
        print('Bienvenido a la plataforma')
        break
if intentos == MAX_INTENTOS:
    print('Completamente explusado del sistema, ladrón.')
```
